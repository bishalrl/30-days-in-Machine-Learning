# webscrapping of Himalayan times

```python
import pandas as pd
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0(Windows NT 6.3;Win 64 ;x64)Apple WeKit /537.36(KHTML ,like Gecko)Chrome/80.0.3987.162 Safari/537.38'}
webpage=requests.get('https://thehimalayantimes.com/business',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

#extracting the heading

```python
for i in soup.find_all('h3',class_='alith_post_title HideMobile' ):
    print(i.text.strip())
```

## exporting specific  news

```python
for i in soup.find_all('h3',class_='alith_post_title_without_image' ):
    print(i.text.strip())
```
